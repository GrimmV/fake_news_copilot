```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)
pd.set_option('display.max_columns', None)

# Load dataset
df = train

# Basic data inspection
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nColumn information:")
print(df.info())

# Handle missing values
print("\nMissing values before cleaning:")
print(df.isnull().sum())
df = df.dropna(subset=['label', 'speaker party'])  # Drop rows with missing critical data

# 1. Label Distribution Analysis
plt.figure(figsize=(10, 6))
label_dist = df['label'].value_counts().sort_index()
sns.barplot(x=label_dist.index, y=label_dist.values, palette='viridis')
plt.title('Distribution of Truthfulness Labels')
plt.xlabel('Truth Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Political Party Analysis
# Party distribution
plt.figure(figsize=(8, 5))
party_dist = df['speaker party'].value_counts()  # Fix typo if needed
sns.barplot(x=party_dist.index, y=party_dist.values, palette='coolwarm')
plt.title('Statements Distribution by Political Party')
plt.xlabel('Political Party')
plt.ylabel('Number of Statements')
plt.yscale("log")
plt.xticks(rotation=45)
plt.show()

# Label distribution by party
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='speaker party', hue='label', palette='viridis')
plt.title('Truthfulness Distribution by Political Party')
plt.xlabel('Political Party')
plt.ylabel('Count')
plt.legend(title='Truth Rating')
plt.yscale("log")
plt.xticks(rotation=45)
plt.show()

# 3. Subject Analysis
# Explode subjects and analyze
subjects_exp = df.explode('subjects')
top_subjects = subjects_exp['subjects'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_subjects.values, y=top_subjects.index, palette='magma')
plt.title('Top 10 Most Common Subjects')
plt.xlabel('Count')
plt.ylabel('Subject')
plt.xticks(rotation=45)
plt.show()

# Subject truth distribution
top_5_subjects = top_subjects.head(5).index
subject_label_dist = subjects_exp[subjects_exp['subjects'].isin(top_5_subjects)]
subject_label_dist = subject_label_dist.groupby(['subjects', 'label']).size().unstack()

subject_label_dist.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Truth Distribution for Top 5 Subjects')
plt.xlabel('Subject')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Truth Rating')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()

# 4. Speaker Analysis
top_speakers = df['speaker'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_speakers.values, y=top_speakers.index, palette='rocket')
plt.title('Top 10 Most Frequent Speakers')
plt.xlabel('Number of Statements')
plt.ylabel('Speaker')
plt.xticks(rotation=45)
plt.show()

# 5. Truth Counts Analysis
truth_columns = ['barely true counts', 'false counts', 'half true counts', 
                'mostly true counts', 'pants on fire counts']

# Correlation between truth counts
corr_matrix = df[truth_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Truth Count Categories')
plt.xticks(rotation=45)
plt.show()

# Average truth counts by party
party_truth = df.groupby('speaker party')[truth_columns].mean().T

party_truth.plot(kind='bar', colormap='coolwarm')
plt.title('Average Truth Counts by Political Party')
plt.xlabel('Truth Category')
plt.ylabel('Average Count')
plt.legend(title='Political Party')
plt.xticks(rotation=45)
plt.yscale("log")
plt.tight_layout()
plt.show()

# 6. Context Analysis (Basic)
context_dist = df['context'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=context_dist.values, y=context_dist.index, palette='plasma')
plt.title('Top 10 Most Common Contexts')
plt.xlabel('Count')
plt.ylabel('Context Type')
plt.xticks(rotation=45)
plt.show()

# Additional Statistics
print("\nKey Statistics:")
print(f"Total statements analyzed: {len(df)}")
print(f"Number of unique speakers: {df['speaker'].nunique()}")
print(f"Number of unique subjects: {subjects_exp['subjects'].nunique()}")
print("\nTruth counts statistics:")
print(df[truth_columns].describe())
```
